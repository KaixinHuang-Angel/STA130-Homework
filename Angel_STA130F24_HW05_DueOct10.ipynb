    "imagine a world where the null hypothesis is true" when calculating a p-value?
    
    We similate tests under the expexctation that our null hypothesis is true, and from those tests we get the sampling distribution of the data under the null hypothesis. From the sampling distribution, if our observed data is very different from what we expected under the null hypothesis, it means our result is very unusual if the null hypothesis is true, which gives us evidence against the null hypothesis. the p-value shows exactly how unusual our data it is.

In simple terms, the p-value shows how compatible our observed data is with the imagined world where the null hypothesis is true, and a smaller p-value states that is is not compatible, suggesting that the null hypothesis is false.

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
n_couples = 124
observed_right_tilts = 80

# Simulate 10,000 trials of 50/50 coin flips for 124 couples
simulations = np.random.binomial(n_couples, 0.5, 10000)

np.random.seed(1)

# Calculate p-value: proportion of simulations where the result is >= 80
p_value = np.mean(simulations >= observed_right_tilts)
print(f'P-value: {p_value}')

# Plot the simulation
plt.hist(simulations, bins=30, color='skyblue', density=True)
plt.axvline(observed_right_tilts, color='red', linestyle='--', label=f'Observed = {observed_right_tilts}')
plt.legend()
plt.title('Simulation of Right Head Tilts')
plt.xlabel('Number of Right Tilts')
plt.ylabel('Density')

# Save as PNG
plt.savefig('right_head_tilts_simulation.png', format='png')

# Display the plot
plt.show()
```

        Alternative hypothesis: Students are not guessing randomly and can indeed determine between the two tea making methods.
        
    Quantitative Analysis
        In the code used to analyse the null hypothesis, we generated 1000 samples with a sample size of 80 and a probability of success of 0.5. For each of these samples, we record what is the percentage of success, which represents the percentage of people that guessed corectly. we then visualize this using a histogram and add the line representing our observed proportion, the 49 out of 80 students who are able state correctly which is poured first. We calculate the p-value using the line p_value = np.mean(sample_proportions >= observed_proportion).     Findings and Discussion
       Looking at the p-value we got, 0.0320, according to the table given in question5, our p-value is in between 0.05 and 0.01, which gives moderate evidence against the null hypothesis.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n_samples = 1000  # Number of simulated samples
sample_size = 80  # Size of each sample
p0 = 0.5  # Null hypothesis proportion

np.random.seed(1)

# Simulate samples
sample_proportions = np.random.binomial(sample_size, p0, n_samples) / sample_size

# Observed proportion
observed_proportion = 49 / 80  # From your data

# Plotting the sample distribution
plt.figure(figsize=(10, 6))
plt.hist(sample_proportions, bins=30, color='lightblue', edgecolor='black', alpha=0.7, density=True)
plt.axvline(observed_proportion, color='red', linestyle='--', label='Observed Proportion (61.25%)')
plt.title('Sample Distribution of Proportions of Correct Identifications')
plt.xlabel('Sample Proportion of Correct Identifications')
plt.ylabel('Density')
plt.legend()
plt.grid()
plt.show()

p_value = np.mean(sample_proportions >= observed_proportion)
print(f"P-value: {p_value:.4f}")
```
